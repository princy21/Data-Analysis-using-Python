Import all the modules numpy, pandas, matplotlib,seaborn and KNN

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sb
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report
from sklearn.linear_model import LogisticRegression
```

Read the zoo.csv file into a Pandas DataSet called: df_zoo


```python
df_zoo = pd.read_csv("zoo.csv")
df_zoo
```

Use the info method to print out some info about df_zoo

```python
df_zoo.info()
```

Read the class.csv file into a Pandas DataSet called df_class

```python
df_class = pd.read_csv("class.csv")
df_class
```

```python
df_class.info()
```

Merge the two DataFrames together.

Use the join method of df_zoo, to merge df_class into df_zoo.

Set the index of df_class to be “Class_Number” before or during the join.

When joining use on=”class_type”

```python
df_zoo= df_zoo.join(df_class.set_index("Class_Number"), on="class_type")
df_zoo
```

Drop unnecessary columns of the new df_zoo in place:

1. catsize

2. Number_Of_Animal_Species_In_Class

3. Animal_Names

```python
df_zoo.drop(["catsize","Number_Of_Animal_Species_In_Class","Animal_Names"], axis = 1, inplace = True)
```

Split the data into training and testing data.

Prepare your X and y:

X: Drop the animal_name, class_type, Class_Type columns.
y: Specify the class_type column.

Use sklearn train_test_split to split the data

```python
x = df_zoo.drop(["animal_name", "class_type", "Class_Type"], axis = 1)
y = df_zoo["class_type"]

x_train, x_test, y_train, y_test= train_test_split(x, y, test_size = 0.3)
```

Create the model and fit it to the training data.

Create an sklearn KNeighborsClassifier model, with k=1.

 Use the fit method to fit it to the training data.

```python
knn = KNeighborsClassifier (n_neighbors = 1)
knn.fit(x_train, y_train)
```

Predict values based on testing data.

Use the predict method to predict values with the X testing data

```python
pred_value = knn.predict(x_test)
```

Print out the classification report using y test data and predictions, to see how accurate the model is.

Insert a markdown cell and make some observations about the accuracy of this model.

```python
print(classification_report(y_test,pred_value))
```

**Observations** 

The accuracy of the data model is 0.94 which is equal to 94%, which is pretty impressive. However, everytime we run the code, the measure of accuracy slightly varies attributing to the randomness of the data being generated. Also, few times the program was run, the accyracy of the model wa 1.0, i.e., 100 per cent.

Repeat steps 10 to 13 using k=3

```python
knn=KNeighborsClassifier(n_neighbors=3)
knn.fit(x_train,y_train)
pred_value=knn.predict(x_test)
print(classification_report(y_test,pred_value))
```

**Observations:** 
    
    We see that the accuracy of the model is 0.90 and lies between 0.84 and 1.00. Since the value of the k did not increase considerably and because it is only 3, we still find that the accuracy of the data is 100%, sometime.
